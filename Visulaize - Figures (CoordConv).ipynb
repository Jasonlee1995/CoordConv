```python
import collections, csv, itertools, os

import numpy as np
import matplotlib.pyplot as plt
```

```python
padded_train, padded_test = './data/padded_train.csv', './data/padded_test.csv'
zero_train, zero_test = './data/zero_train.csv', './data/zero_test.csv'
canvas_size = 64

csvs = [padded_train, padded_test, zero_train, zero_test]


coords = []
for csv_file in csvs:
    statistics = np.array([[0] * canvas_size for _ in range(canvas_size)])
    with open(csv_file, newline='') as f:
        reader = csv.reader(f)
        for row in reader:
            _, x, y = row
            x, y = int(x), int(y)
            statistics[y, x] = 255
    coords.append(statistics)
    

plt.figure(figsize=(20, 8))
plt.axis('off')

for i in range(len(coords)):
    plt.subplot(1, len(coords), i+1)
    img = plt.imshow(coords[i]); img.axes.get_xaxis().set_visible(False); img.axes.get_yaxis().set_visible(False)
```

```python
train_test = ['train', 'test']
data_type = ['padded', 'zero']
mode = ['coord']
sigmoid_loss = ['sig', 'no']
code_paper = ['paper']

infos = collections.OrderedDict()
for info in itertools.product(train_test, data_type, mode, sigmoid_loss, code_paper):
    tag = '_'.join(list(info))
    infos[tag] = []
    
    
fig_dir = './_figures/'

for file in sorted(os.listdir(fig_dir)):
    for tag in infos:
        if file.startswith(tag):
            infos[tag].append(fig_dir + file)
```
