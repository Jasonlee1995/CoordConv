```python
import generate_data

import numpy as np
import matplotlib.pyplot as plt
```

```python
train_num, test_num = 3000, 1000
canvas_size = 64
```

```python
pad = 0


trains, tests = [], []
for i in range(5):
    train_coords, test_coords, _, _ = generate_data.train_test(i, canvas_size, pad, train_num, test_num)
    
    train_statistics = np.array([[0] * canvas_size for _ in range(canvas_size)])
    test_statistics = np.array([[0] * canvas_size for _ in range(canvas_size)])
    
    for x, y in train_coords: train_statistics[y, x] = 255
    for x, y in test_coords: test_statistics[y, x] = 255
    
    trains.append(train_statistics); tests.append(test_statistics)


total = trains + tests


plt.figure(figsize=(20, 8))
plt.rcParams['axes.titlesize'] = 20
plt.axis('off')

for i in range(1, len(total)+1):
    plt.subplot(2, len(total)//2, i)
    
    img = plt.imshow(total[i-1]); img.axes.get_xaxis().set_visible(False); img.axes.get_yaxis().set_visible(False)
    if i > len(total)//2: plt.gca().set_title('Test')
    else: plt.gca().set_title('Train')
```
